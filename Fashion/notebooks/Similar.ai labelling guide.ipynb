# Similar.AI Labelling Guide

For the dataset provided, we ask that Similar.ai only label individuals identified by a red bounded box. For these individuals, we already have some upper body labels and regions to localise when training our model. This will enchance the metric learning process for mapping Similar.ai labels to existing StreetStyle labels. So far we have provided 23,908 images for labelling. **Labels required** 

We have labels for most of the upper body clothing types. What we need is item type labelling for the lower body as well as color and market, as well as some misc item labels.

So for example, lets look at the labels for this picture:

````python
Image(filename='8c5663f4fd701e5d5c6b08e4821af07f_761585291335355200_1146068377.jpg',width=250,height=200) 
````

First, we only want to look at the person with the bounding box, we're going to ignore everybody else. We have upper body labels for this image (so, for example, color=black, item_type=sweater) so we need the lower body labels. For this image, we could have ```[market_woman, trousers, blue, gloves, bag]```.

Any label ommited, eg ```belt```, we assume to mean ```belt_none``` so you don't have to worry about labelling things that aren't there.

Because of the way the labelling is set up at on platform.ai at the moment, we can only assign colour to the entire image. Please select whatever colour best represents the lower body item type (e.g. trousers, dress, onesies) even if it matches the upper body colour. In the future we can look at doing more granular colour (e.g. for bags, hats, etc) although that would require more labelling projects. Platform.ai clusters images based on the labels to make your job easier, so we don't want to include all the labels in one project.

Taking labels from your notion document:

We lack labels the item types: ```skirt```, ```hosiery```, ```trousers```, ```lingerie```, ```lounge/sleepwear```, ```swimwear```, ```onesies```, ```belts```, ```hair accessories```, ```jewellery```, ```shoes```, ```bags``` and ```gloves```.

We have no labels for market.

We have the same colour labels as you do, so just use the colours you have listed with the addition of a label for multicoloured items.

For everything else listed in the notion document that we haven't listed above, we have labels for.

The labels already exist within the platform.ai projects so you won't have to remember this list.

The streestyle dataset additionally includes a ```clothing_pattern``` label (solid, graphics, striped, floral, plaid and spotted). We don't need to work with this right now, but if you want to include it (for the item types we don't have labels for, as listed above) that can also work.

If you do decide to work with clothing pattern, the labels are already set up in the project. Please select the clothing pattern that best represents the lower body item type (ignore bags etc).
